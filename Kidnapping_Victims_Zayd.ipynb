```python
import numpy as np
import os
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
sb.set()
from sklearn.model_selection import train_test_split
```

<h1>Sample Collection<h1>

```python
k_data = pd.read_csv('Data/archive/39_Kidnapping_and_abduction.csv')
k_data.head()
```

```python
total_k = pd.DataFrame(k_data[['Area_Name','Year', 'K_A_Cases_Reported']])
total_k.head()
```

```python
r_data = pd.read_csv('Data/archive/42_Cases_under_crime_against_women.csv')
r_data.head()
```

```python
total_r = pd.DataFrame(r_data[['Area_Name', 'Year', 'Total_Cases_for_Trial']])
total_r.head()
```

```python
combined_df = pd.merge(total_k, total_r, on=['Area_Name', 'Year'])
combined_df.head()
```

<h1>Data Preparation<h1>

```python
combined_clean=combined_df.copy()
```

```python
combined_clean.rename(columns = {'K_A_Cases_Reported' : 'Abduction_cases', 'Total_Cases_for_Trial' : 'Rape_cases'}, inplace = True)
combined_clean.columns = combined_clean.columns.str.upper()
combined_clean.head()
```

```python
combined_clean = combined_clean.dropna(subset=['ABDUCTION CASES', 'RAPE_CASES'])
```

```python
Q1 = combined_clean['RAPE_CASES'].quantile(0.25)
Q3 = combined_clean['RAPE_CASES'].quantile(0.75)
IQR = Q3 - Q1

lbound = Q1- 1.5 * IQR
ubound = Q3 + 1.5 * IQR

combined_clean = combined_clean[(combined_clean['RAPE_CASES'] >= lbound) & (combined_clean['RAPE_CASES'] <= ubound)]
```

```python
combined_clean.info()
```

```python
abduct = pd.DataFrame(combined_clean['ABDUCTION CASES'])
rape = pd.DataFrame(combined_clean['RAPE_CASES'])
```

<h1>Algorithmic Optmisation<h1>

<h2>Linear Regression<h2>

```python
from sklearn.linear_model import LinearRegression
```

```python
rape_train, rape_test = train_test_split(rape, test_size=0.25)
abduct_train, abduct_test = train_test_split(abduct, test_size=0.25)
```

```python
rape_train.head()
```

```python
rape_test.head()
```

```python
linreg = LinearRegression()
```

```python
linreg.fit(abduct_train, rape_train)
```

```python
print('Intercept \t: b = ', linreg.intercept_)
print('Coefficients \t: a = ', linreg.coef_)
```

```python
rape_train_pred = linreg.predict(abduct_train)

f = plt.figure(figsize=(16, 8))
plt.scatter(abduct_train, rape_train)
plt.scatter(abduct_train, rape_train_pred, color = "r")
plt.show()
```

```python
print("Explained Variance (R^2) \t:", linreg.score(abduct_train, rape_train))

def mean_sq_err(actual, predicted):
    return np.mean(np.square(np.array(actual) - np.array(predicted)))


mse = mean_sq_err(rape_train, rape_train_pred)
print("Mean Squared Error (MSE) \t:", mse)
print("Root Mean Squared Error (RMSE) \t:", np.sqrt(mse))
```

```python
rape_test_pred = linreg.predict(abduct_test)

f = plt.figure(figsize=(16, 8))
plt.scatter(abduct_test, rape_test, color = "green")
plt.scatter(abduct_test, rape_test_pred, color = "red")
plt.show()
```

```python
print("Explained Variance (R^2) \t:", linreg.score(abduct_test, rape_test))
mse = mean_sq_err(rape_test, rape_test_pred)
print("Mean Squared Error (MSE) \t:", mse)
print("Root Mean Squared Error (RMSE) \t:", np.sqrt(mse))
```

```python
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
X_train = abduct_train.values.reshape(-1, 1)
X_test = abduct_test.values.reshape(-1, 1) 
```

```python
neuralnetwork = Sequential([
    Dense(10, activation='relu', input_shape=(1,)),  # Input layer with 10 neurons, change depending on complexity
    Dense(1)  # Output layer with 1 neuron
])

# Compile the model
neuralnetwork.compile(optimizer='adam', loss='mean_squared_error')
```

```python
neuralnetwork.fit(X_train, rape_train, epochs=100, validation_split=0.2)
```

```python
rape_train_pred = neuralnetwork.predict(abduct_train)

f = plt.figure(figsize=(16, 8))
plt.scatter(abduct_train, rape_train)
plt.scatter(abduct_train, rape_train_pred, color = "r")
plt.show()
```

```python
rape_test_pred = neuralnetwork.predict(abduct_test)

f = plt.figure(figsize=(16, 8))
plt.scatter(abduct_test, rape_test, color = "green")
plt.scatter(abduct_test, rape_test_pred, color = "red")
plt.show()
```

As such, the total number of kidnapping and abduction cases may be a important predictor of the number of rape cases reported.
    
Three machine learning models were utilised for algorithmic optimisation. A linear regression model, a KNN model and a neural network. Each of the three machine learning models had a high mean squared error, and as a result, had difficulty accuratley predicting the total number of sexual assault cases in response to the total number of kidnapping and abduction cases. THe linear regression model had the best prediction, and had the lowest mean squared error, compared to the KNN model and the neural network. 
    
Overall, none of the three models appears to be accurate in predicting the total number of sexual assault cases, despite the high correlation. This could be due to a number of reasons, such as a large number of anomalies which need to be filtered out further. Another possibility is that the relationship between the data is not linear, it could be quadratic or another form of clustering<p>
